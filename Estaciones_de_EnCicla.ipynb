# Transformación de Datos Geoespaciales: Estaciones de EnCicla



Este repositorio contiene el proceso de transformación de datos geoespaciales del dataset de estaciones de EnCicla en Medellín. La información original, almacenada en formato .csv, fue convertida a un archivo Shapefile compatible con ArcGIS, permitiendo la creación de una capa geoespacial para el análisis y visualización en esta plataforma.

```python
import geopandas as gpd
import matplotlib.pyplot as plt
import geopandas as gpd
import pandas as pd
import zipfile
import os
```

```python
from google.colab import files
uploaded = files.upload()
```

```python
df = pd.read_csv("Estaciones_EnCicla_AMVA_20241111.csv")
```

```python
df.head()
```

```python
df[['lat', 'lon']] = df['Georeferenciación'].str.split(';', expand=True)
df['lat'] = df['lat'].str.replace(',', '.').astype(float)
df['lon'] = df['lon'].str.replace(',', '.').astype(float)

# Convertir el DataFrame a un GeoDataFrame
gdf = gpd.GeoDataFrame(df, geometry=gpd.points_from_xy(df['lon'], df['lat']))

# Establecer el CRS (por ejemplo, WGS84)
gdf.set_crs(epsg=4326, inplace=True)

# Eliminar la columna Georeferenciacion
gdf = gdf.drop(columns=['Georeferenciación', 'lat', 'lon'])
```

```python
gdf.head()
```

```python
gdf.to_file("/content/Estaciones_ENCICLA.shp")
```

```python
shapefile_dir = "/content/Estaciones_ENCICLA"  # Nombre base del shapefile
zip_filename = "/content/Estaciones_ENCICLA.zip"

with zipfile.ZipFile(zip_filename, 'w') as zipf:
    for ext in ['.shp', '.shx', '.dbf', '.prj']:
        zipf.write(shapefile_dir + ext, os.path.basename(shapefile_dir + ext))
```
